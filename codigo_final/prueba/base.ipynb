## Main base

```python
import warnings
# IMPORTANTE! Importa tu nuevo archivo
from arma_pipeline import Pipeline

# El punto de entrada DEBE estar protegido por esta guarda
if __name__ == '__main__':
    warnings.filterwarnings("ignore")

    VERBOSE_MODE = False
    NUM_SAMPLES = 300
    PHI_COEFFS = [0.8, -0.2]
    THETA_COEFFS = [0.5]
    MODELO_ARMA = f"ARMA({len(PHI_COEFFS)},{len(THETA_COEFFS)})"
    SIGMA_RUIDO = 1.5
    DIST_RUIDO = 't-student'

    # Instanciar y ejecutar el pipeline
    # Ahora Python sabe exactamente dnde encontrar el cdigo de la clase Pipeline
    pipeline = Pipeline(
        model_type=MODELO_ARMA,
        phi=PHI_COEFFS,
        theta=THETA_COEFFS,
        sigma=SIGMA_RUIDO,
        noise_dist=DIST_RUIDO,
        n_samples=NUM_SAMPLES,
        seed=42,
        verbose=VERBOSE_MODE 
    )

    pipeline.execute()
```
