# AplicaciÃ³n

```python
from pipeline import Pipeline

VERBOSE_MODE = False

# NÃºmero de puntos a simular (sin contar el "burn-in")
NUM_SAMPLES = 300

# ParÃ¡metros del proceso ARMA
# phi son los coeficientes AR, theta son los coeficientes MA
PHI_COEFFS = [0.8, -0.2]  # Un AR(2)
THETA_COEFFS = [0.5]      # Un MA(1)
MODELO_ARMA = f"ARMA({len(PHI_COEFFS)},{len(THETA_COEFFS)})"

# DesviaciÃ³n estÃ¡ndar y tipo de ruido ('normal' o 't-student')
SIGMA_RUIDO = 1.5
DIST_RUIDO = 't-student' # t-student tiene colas mÃ¡s pesadas (mÃ¡s outliers)

# ----------------------------------------

# Instanciar y ejecutar el pipeline con la configuraciÃ³n definida
pipeline = Pipeline(
    model_type=MODELO_ARMA,
    phi=PHI_COEFFS,
    theta=THETA_COEFFS,
    sigma=SIGMA_RUIDO,
    noise_dist=DIST_RUIDO,
    n_samples=NUM_SAMPLES,
    seed=42,
    verbose=VERBOSE_MODE 
)

pipeline.execute()
```
